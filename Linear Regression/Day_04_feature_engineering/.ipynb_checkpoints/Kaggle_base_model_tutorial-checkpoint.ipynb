```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('./ks-projects-201801.csv', parse_dates=['deadline','launched'])
data.head()
```

```python
pd.unique(data.state)
```

```python
data.groupby('state')['ID'].count()
```

```python
data = data.query('state != "live"')

data = data.assign(outcome=(data['state'] == 'successful').astype(int))
data.head()
```

```python
data = data.assign(hour=data.launched.dt.hour,
                  day=data.launched.dt.day,
                  month=data.launched.dt.month,
                  year=data.launched.dt.year)
data.head()
```

```python
from sklearn.preprocessing import LabelEncoder

cat_features = ['category', 'currency', 'country']
encoder = LabelEncoder()

encoded_features = data[cat_features].apply(encoder.fit_transform)
encoded_features.head()
```

```python

new_data = data[['goal', 'hour','day', 'month', 'year', 'outcome']].join(encoded_features)
new_data.head()
```

```python
frac = 0.1
size = int(len(new_data)*frac)

train_data = new_data[:-2*size]
valid_data = new_data[-2*size:-size]
test_data = new_data[-size:]
```

```python
for i in [train_data,valid_data,test_data]:
    print(f'Mean = {i.outcome.mean():0.4f}')
```

```python
import lightgbm as lgb

feature_col = train_data.columns.drop('outcome')

dtrain = lgb.Dataset(train_data[feature_col], label=train_data['outcome'])
dvalid = lgb.Dataset(valid_data[feature_col], label=valid_data['outcome'])

param = {'num_leaves': 64, 'objective': 'binary'}
param['metric'] = 'auc'
num_round = 1000
bst = lgb.train(param, dtrain, num_round, valid_sets=[dvalid], early_stopping_rounds=10, verbose_eval=False)
```

```python
from sklearn import metrics
ypred = bst.predict(test_data[feature_col])
score = metrics.roc_auc_score(test_data['outcome'], ypred)

print(f"Test AUC score: {score}")
```
